```python
def integer_params_decorator(v):
    def wrapper(*args):
        if not all(map(lambda x: isinstance(x, int), args[1:])):
            raise TypeError
        res = v(*args)
        return res
    return wrapper


def integer_params(cls):
    methods = {k: v for k, v in cls.__dict__.items() if callable(v)}
    for k, v in methods.items():
        setattr(cls, k, integer_params_decorator(v))
    return cls


@integer_params
class Vector:
    def __init__(self, *args):
        self.__coords = list(args)

    def __getitem__(self, item):
        return self.__coords[item]

    def __setitem__(self, key, value):
        self.__coords[key] = value

    def set_coords(self, *coords, reverse=False):
        c = list(coords)
        self.__coords = c if not reverse else c[::-1]


vector = Vector(1, 2)
print(vector[0])
print(vector[1])
vector[1] = 54
print(vector[1])
```

```python
class TriangleChecker:
    def __init__(self, a, b, c):
        self.sides = a, b, c

    def is_triangle(self):
        if any(map(lambda x: type(x) not in (int, float) or x <= 0, self.sides)):
            return 1
        if self.sides[0] + self.sides[1] < self.sides[2] or self.sides[1] + self.sides[2] < self.sides[0] \
                or self.sides[0] + self.sides[2] < self.sides[1]:
            return 2
        else:
            return 3


tr = TriangleChecker(3, 4, 5)
print(tr.is_triangle())
```

```python
class Factory:
    def build_sequence(self):
        sequence = []
        return sequence

    def build_number(self, string):
        self.string = string
        return float(string)

class Loader:
    def parse_format(self, string, factory):
        seq = factory.build_sequence()
        for sub in string.split(','):
            item = factory.build_number(sub)
            seq.append(item)
        return seq

ld = Loader()
res = ld.parse_format('4, 5, -6.5', Factory())
print(res)
```

```python
class Server:
    COUNTER = 1

    def __init__(self):
        self.ip = Server.COUNTER
        Server.COUNTER += 1
        self.buffer_server = []
        self.router = None  # чтобы в другом классе вызвать этот локальный атрибут экземпляра (связывание двух классов)

    def send_data(self, data):
        self.router.buffer_router.append(data)

    def get_data(self):
        num = self.buffer_server[:]
        self.buffer_server.clear()
        return num

    def get_ip(self):
        return self.ip


class Router:

    def __init__(self):
        self.connect_servers = {}
        self.buffer_router = []

    def link(self, server):
        if server not in self.connect_servers:
            self.connect_servers.update({server.ip: server})
            server.router = self

    def unlink(self, server):
        del self.connect_servers[server.ip]

    def send_data(self):
        for i in self.buffer_router:
            if i.ip in self.connect_servers:
                self.connect_servers[i.ip].buffer_server.append(i)
        self.buffer_router.clear()


class Data:
    def __init__(self, data, ip):
        self.data = data
        self.ip = ip


router = Router()
sv_from = Server()
sv_from2 = Server()
router.link(sv_from)
router.link(sv_from2)
router.link(Server())
router.link(Server())
sv_to = Server()
router.link(sv_to)
sv_from.send_data(Data('Hello', sv_to.get_ip()))
sv_from2.send_data(Data('Hello', sv_to.get_ip()))
sv_to.send_data(Data('Hi', sv_from.get_ip()))
router.send_data()
msg_lst_from = sv_from.get_data()
msg_lst_to = sv_to.get_data()
print(router.connect_servers)
print(router.buffer_router)
print(len(sv_from.buffer_server))
assert len(router.buffer_router) == 0, 'после отправки сообщений буфер роутора должен очищаться'
assert len(sv_from.buffer_server) == 0, 'после отправки сообщений буфер сервера должен очищаться'
assert len(msg_lst_to) == 2, 'метод get_data вернул неверное число пакетов'
print(len(msg_lst_to))
assert msg_lst_from[0].data == 'Hi' and msg_lst_to[0].data == 'Hello', 'данные не прошли по сети, классы не функционируют должным образом'
assert hasattr(router, 'buffer_router') and hasattr(sv_to, 'buffer_server'), 'в объектах класса отсутстует локальный атрибут buffer'
router.unlink(sv_to)
sv_from.send_data(Data('Hello', sv_to.get_ip()))
router.send_data()
msg_lst_to = sv_to.get_data()
assert msg_lst_to == [], 'метод гет дата вернул неверные данные'
```

```python
class WindowDlg:

    def __init__(self, title, width, height):
        self.__title = title
        if self.__check_num(width) and self.__check_num(height):
            self.__width = width
            self.__height = height

    def show(self):
        print(f'{self.__title}: {self.__width}, {self.__height}')

    @staticmethod
    def __check_num(num):
        return isinstance(num, int) and 0 <= num <= 10000

    @property
    def width(self):
        return self.__width
    
    @width.setter
    def width(self, width):
        if self.__check_num(width) and self.__width != width:
            self.__width = width
            self.show()

    @property
    def height(self):
        return self.__height

    @height.setter
    def height(self, height):
        if self.__check_num(height) and self.__height != height:
            self.__height = height
            self.show()


wd = WindowDlg('Table', 122, 10)
wd.width = 123
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def add_obj(self, obj):
        if self.head is None:
            self.head = obj
        else:
            n = self.head
            while n.get_next() is not None:
                n = n.get_next()
            n.set_next(obj)

    def remove_obj(self):
        a = []
        if self.head is None:
            return
        if self.head.get_next() is None:
            a.append(self.head)
            self.head = None
            return a[0]
        else:
            n = self.head
            while n.get_next().get_next() is not None:
                n = n.get_next()
            n.set_next(None)
            return n

    def get_data(self):
        total_list = []
        h = self.head
        while h:
            total_list.append(h.data)
            h = h.next
        return total_list


class ObjList:
    def __init__(self, data):
        self.__next = None
        self.__data = data

    def set_next(self, obj):
        self.__next = obj

    def get_next(self):
        return self.__next

    def set_data(self, data):
        self.__data = data

    def get_data(self):
        return self.__data

    next = property(get_next, set_next)
    data = property(get_data, set_data)


ls = LinkedList()
top = ObjList('данные 1')
ls.add_obj(top)
ls.add_obj(ObjList('данные 2'))
ls.add_obj(ObjList('данные 3'))
ls.remove_obj()
res = ls.get_data()
print(ls.get_data())
assert res == ['данные 1', 'данные 2']
assert ls.head == top

h = ls.head
while h:
    res = h.data
    h = h.next
s = LinkedList()
top = ObjList('данные 1')
s.add_obj(top)
s.remove_obj()

assert s.get_data() == []
n = 0
h = s.head
while h:
    h = h.next
    n += 1
assert n == 0

s = LinkedList()
top = ObjList('name_1')
s.add_obj(top)
obj = s.remove_obj()
print(obj, top)
assert obj == top
```

```python
class RadiusVector2d:
    MIN_COORD = -100
    MAX_COORD = 1024

    def __init__(self, x, y):
        if self._check(x):
            self.__x = x
        else:
            self.__x = 0
        if self._check(y):
            self.__y = y
        else:
            self.__y = 0

    @property
    def x(self):
        return self.__x

    @x.setter
    def x(self, x):
        if self._check(x):
            self.__x = x
        else:
            self.__x = 0

    @property
    def y(self):
        return self.__y

    @y.setter
    def y(self, y):
        if self._check(y):
            self.__y = y
        else:
            self.__y = 0

    @staticmethod
    def norm2(vector):
        return vector.x ** 2 + vector.y ** 2

    def _check(self, data):
        if type(data) in (int, float) and self.MIN_COORD <= data <= self.MAX_COORD:
            return True


v1 = RadiusVector2d(1123123, 2)
print(RadiusVector2d.norm2(v1))
```

```python
import copy


class ListMath:
    def __init__(self, *args):
        if len(args) > 0:
            self.args = list(filter(lambda x: type(x) in (int, float), args[0]))
        else:
            self.args = []
        self.lst_math = copy.deepcopy(self.args)

    def __add__(self, other):
        n = copy.deepcopy(other)
        s = copy.deepcopy(self.lst_math)
        return ListMath(list(map(lambda x: x + n, s)))

    def __radd__(self, other):
        return self.__add__(other)

    def __sub__(self, other):
        n = copy.deepcopy(other)
        s = copy.deepcopy(self.lst_math)
        return ListMath(list(map(lambda x: x - n, s)))

    def __rsub__(self, other):
        n = copy.deepcopy(other)
        s = copy.deepcopy(self.lst_math)
        return ListMath(list(map(lambda x: n - x, s)))

    def __mul__(self, other):
        n = copy.deepcopy(other)
        s = copy.deepcopy(self.lst_math)
        return ListMath(list(map(lambda x: x * n, s)))

    def __rmul__(self, other):
        return self.__mul__(other)

    def __truediv__(self, other):
        n = copy.deepcopy(other)
        s = copy.deepcopy(self.lst_math)
        return ListMath(list(map(lambda x: x / n, s)))

    def __rtruediv__(self, other):
        n = copy.deepcopy(other)
        s = copy.deepcopy(self.lst_math)
        return ListMath(list(map(lambda x: n / x, s)))
```

```python
import math


class PositiveValue:
    @classmethod
    def verify_num(cls, num):
        if num < 0:
            raise ValueError('Длины д.б. только > 0')

    def __set_name__(self, owner, name):
        self.name = '__' + name

    def __get__(self, instance, owner):
        return getattr(instance, self.name)

    def __set__(self, instance, value):
        self.verify_num(value)
        setattr(instance, self.name, value)


class Triangle:
    a = PositiveValue()
    b = PositiveValue()
    c = PositiveValue()

    def __init__(self, a, b, c):
        if self.verify_sides(a, b, c):
            self.a = a
            self.b = b
            self.c = c

    def __call__(self):
        p = (self.a + self.b + self.c)//2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

    def __len__(self):
        return int(self.a + self.b + self.c)

    @staticmethod
    def verify_sides(a, b, c):
        if a < b + c and b < a + c and c < a + b:
            return True
        else:
            raise ValueError('нельзя образовать треугол')
```

```python
class Vector:
    def __init__(self, *args):
        self.args = args

    def __add__(self, other):
        l = []
        if isinstance(other, Vector):
            if len(self.args) != len(other.args):
                raise ArithmeticError('args doesnt have the same length')
            for i in range(len(self.args)):
                l.append(self.args[i] + other.args[i])
            return Vector(*l)
        return Vector(*list(map(lambda x: x + other, self.args)))

    def __sub__(self, other):
        l = []
        if isinstance(other, Vector):
            if len(self.args) != len(other.args):
                raise ArithmeticError('args doesnt have the same length')
            for i in range(len(self.args)):
                l.append(self.args[i] - other.args[i])
            return Vector(*l)
        return Vector(*list(map(lambda x: x - other, self.args)))

    def __mul__(self, other):
        l = []
        if isinstance(other, Vector):
            if len(self.args) != len(other.args):
                raise ArithmeticError('args doesnt have the same length')
            for i in range(len(self.args)):
                l.append(self.args[i] * other.args[i])
            return Vector(*l)
        return Vector(*list(map(lambda x: x * other, self.args)))

    def __eq__(self, other):
        l = []
        if len(self.args) != len(other.args):
            raise ArithmeticError('args doesnt have the same length')
        for i in range(len(self.args)):
            l.append(self.args[i] == other.args[i])
        return all(map(lambda x: x is True, l))
```

```python
#Паттерн слушатель


class Data:
    def __init__(self, temp, press, wet):
        self.temp = temp
        self.press = press
        self.wet = wet


class Observer:
    def update(self, data):
        pass

    def __hash__(self):
        return hash(id(self))


class Subject:
    def __init__(self):
        self.__observers = {}
        self.__data = None

    def add_observer(self, observer):
        self.__observers[observer] = observer

    def remove_observer(self, observer):
        if observer in self.__observers:
            self.__observers.pop(observer)

    def __notify_observer(self):
        for ob in self.__observers:
            ob.update(self.__data)

    def change_data(self, data):
        self.__data = data
        self.__notify_observer()


class TemperatureView(Observer):
    def update(self, data):
        print(f'Текущая температура {data.temp}')


class PressureView(Observer):
    def update(self, data):
        print(f'Текущае давление {data.press}')


class WetView(Observer):
    def update(self, data):
        print(f'Текущая влажность {data.wet}')


sub = Subject()
tv = TemperatureView()
pr = PressureView()
wet = WetView()
sub.add_observer(tv)
sub.add_observer(pr)
sub.add_observer(wet)
```

```python
vector_vlog = []


def decorator(v, vector):
    def wraps(*args):
        res = v(*args)
        vector.append(v.__name__)
        return res
    return wraps


def class_log(vector):
    def wrapper(cls):
        methods = {k: v for k, v in cls.__dict__.items() if callable(v)}
        for k, v in methods.items():
            setattr(cls, k, decorator(v, vector))
        return cls
    return wrapper


@class_log(vector_vlog)
class Vector:
    def __init__(self, *args):
        self.__coords = list(args)

    def __getitem__(self, item):
        return self.__coords[item]

    def __setitem__(self, key, value):
        self.__coords[key] = value

    def get_coords(self):
        return self.__coords


v = Vector(1, 2, 3)
print(v.get_coords())
print(vector_vlog)
```

```python
class Money:
    def __init__(self, value):
        if type(value) not in (int, float):
            raise TypeError
        self._money = value

    @property
    def money(self):
        return self._money

    @money.setter
    def money(self, mon):
        self._money = mon


class MoneyOperators:
    def __add__(self, other):
        if type(other) in (int, float):
            return self.__class__(self.money + other)
        if type(self) != type(other):
            raise TypeError
        return self.__class__(self.money + other.money)

    def __sub__(self, other):
        if type(other) in (int, float):
            return self.__class__(self.money - other)
        if type(self) != type(other):
            raise TypeError
        return self.__class__(self.money - other.money)


class MoneyR(Money, MoneyOperators):
    def __str__(self):
        return f'MoneyR: {self.money}'


class MoneyD(Money, MoneyOperators):
    def __str__(self):
        return f'MoneyD: {self.money}'

```

```python
class Function:
    def __init__(self):
        self._amplitude = 1.0
        self._bias = 0.0

    def __call__(self, x, *args, **kwargs):
        return self._amplitude * self._get_function(x) + self._bias

    def _get_function(self, x):
        raise NotImplementedError('не переопределен метод')

    def __add__(self, other):
        if type(other) not in (int, float):
            raise TypeError

        obj = self.__class__(self)
        obj._bias = self._bias + other
        return obj

    def __mul__(self, other):
        if type(other) not in (int, float):
            raise TypeError

        obj = self.__class__(self)
        obj._amplitude = self._amplitude * other
        return obj


class Linear(Function):
    def __init__(self, *args):
        super().__init__()
        if len(args) == 2:
            self._k = args[0]
            self._b = args[1]
        else:
            self._k = args[0]._k
            self._b = args[0]._b

    def __call__(self, x, *args, **kwargs):
        return self._amplitude * (self._k * self._get_function(x) + self._b) + self._bias

    def _get_function(self, x):
        return x


f = Linear(1, 0.5)
f1 = f + 5
print(f1.__dict__)
print(f1(0))
```

```python
def pick_peaks(array):
    result = {'pos': [], 'peaks': []}
    k, m, n = 0, 0, 1
    l = {}
    if len(array) == 0 or array.count(array[0]) == len(array):
        return result
    while array[m] >= array[n]:
        m += 1
        n += 1
        k += 1
    while k != len(array)-1:
        while k + 1 < len(array) and array[k] <= array[k+1]:
            if array[k] == array[k + 1]:
                l.setdefault(array[k], k)
            k += 1
        if k + 1 < len(array) and array[k] > array[k+1] and array[k] >= array[k-1]:
            if array[k] == array[k-1]:
                result['pos'] = result.get('pos', []) + [l[array[k]]]
            else:
                result['pos'] = result.get('pos', []) + [k]
            result['peaks'] = result.get('peaks', []) + [array[k]]
        while k + 1 < len(array) and array[k] >= array[k+1]:
            k += 1
    return result
```
